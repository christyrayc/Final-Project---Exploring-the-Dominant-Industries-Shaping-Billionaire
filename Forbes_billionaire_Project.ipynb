```python
# Group the data by 'country / territory' and calculate the total net worth for each country
country_wealth = df.groupby('country / territory')['net_worth (in billion $)'].sum()

# Group the data by 'industry' and calculate the total net worth for each industry
industry_wealth = df.groupby('industry')['net_worth (in billion $)'].sum()

# Print the results
print("Wealth Distribution Across Countries:")
print(country_wealth)

print("\nWealth Distribution Across Industries:")
print(industry_wealth)
```

```python
# # Analyzing wealth distribution across countries and industries

# Plot the wealth distribution across countries
plt.figure(figsize=(10, 6))
country_wealth.plot(kind='bar')
plt.title('Wealth Distribution Across Countries')
plt.xlabel('Country / Territory')
plt.ylabel('Total Net Worth (in billion $)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# # Identifying the top sources of wealth and the most successful industries

# In[33]:


# Group the data by 'source' and calculate the total net worth for each source
source_wealth = df.groupby('source')['net_worth (in billion $)'].sum()

# Group the data by 'industry' and calculate the total net worth for each industry
industry_wealth = df.groupby('industry')['net_worth (in billion $)'].sum()

# Find the top sources of wealth
top_sources = source_wealth.nlargest(30)


# Find the most successful industries
most_successful_industries = industry_wealth.nlargest(18)

print("Top Sources of Wealth:")
print(top_sources)

print("\nMost Successful Industries:")
print(most_successful_industries)
```

```python
# Visualize the top sources of wealth
plt.figure(figsize=(10, 6))
top_sources.plot(kind='bar', color='skyblue')
plt.title('Top Sources of Wealth')
plt.xlabel('Source')
plt.ylabel('Total Net Worth (in billion $)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Visualize the most successful industries
plt.figure(figsize=(10, 6))
most_successful_industries.plot(kind='bar', color='lightgreen')
plt.title('Most Successful Industries')
plt.xlabel('Industry')
plt.ylabel('Total Net Worth (in billion $)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Group the data by Continent and Country, then sum the wealth
wealth_by_country = df.groupby(['Continent', 'country / territory'])['net_worth (in billion $)'].sum()

# Sort the wealth by continent and country
sorted_wealth_by_country = wealth_by_country.groupby('Continent', group_keys=False).apply(lambda x: x.sort_values(ascending=False))

# Get the top 5 countries in each continent
top_countries_by_continent = sorted_wealth_by_country.groupby('Continent').head(5)

print("Top 5 Countries by Continent:")
print(top_countries_by_continent)
```

```python
# Group the data by Continent and sum the wealth
wealth_by_continent = df.groupby('Continent')['net_worth (in billion $)'].sum()

# Sort the wealth by continent
sorted_wealth_by_continent = wealth_by_continent.sort_values(ascending=False)

# Iterate through each continent and calculate the count of people
people_count_by_continent = {}
for continent in sorted_wealth_by_continent.index:
    total_wealth = sorted_wealth_by_continent[continent]
    people_count = df[df['Continent'] == continent].shape[0]
    people_count_by_continent[continent] = (people_count, total_wealth)

# Display the count of people and total wealth by continent
for continent, (people_count, total_wealth) in people_count_by_continent.items():
    print(f"Continent: {continent}\t People Count: {people_count}\t Total Wealth: {total_wealth}")
```

```python
# Group the data by Continent and sort by net worth in descending order
grouped_by_continent = df.groupby('Continent', as_index=False).apply(lambda x: x.nlargest(5, 'net_worth (in billion $)'))

# Display the top 5 richest people in each continent
for continent, data in grouped_by_continent.groupby('Continent'):
    print(f"Continent: {continent}")
    print(data[['rank', 'name', 'net_worth (in billion $)']])
    print()
```

```python
import matplotlib.pyplot as plt

# Group the data by Continent and sort by net worth in descending order
grouped_by_continent = df.groupby('Continent', as_index=False).apply(lambda x: x.nlargest(5, 'net_worth (in billion $)'))

# Create subplots for each continent
fig, axes = plt.subplots(nrows=len(grouped_by_continent), figsize=(10, 6 * len(grouped_by_continent)))

# Plot the top 5 richest people for each continent
for i, (continent, data) in enumerate(grouped_by_continent.groupby('Continent')):
    ax = axes[i]
    data[['name', 'net_worth (in billion $)']].plot(kind='bar', x='name', y='net_worth (in billion $)', ax=ax)
    ax.set_title(f'Top 5 Richest People in {continent}')
    ax.set_xlabel('Name')
    ax.set_ylabel('Net Worth (in billion $)')
    ax.legend().remove()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group the data by Continent and Industry, and sum up wealth
wealth_by_continent_industry = df.groupby(['Continent', 'industry'])['net_worth (in billion $)'].sum().reset_index()

# Pivot the data to get it in the right format for plotting
pivot_data = wealth_by_continent_industry.pivot(index='Continent', columns='industry', values='net_worth (in billion $)')

# Create a grouped bar plot
plt.figure(figsize=(12, 6))
sns.set_palette("Set3")
pivot_data.plot(kind='bar', stacked=True)
plt.title('Wealth Distribution by Continent and Industry')
plt.xlabel('Continent')
plt.ylabel('Total Wealth (in billion $)')
plt.xticks(rotation=45)
plt.legend(title='Industry', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.show()
```
